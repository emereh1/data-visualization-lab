```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV into a DataFrame
df = pd.read_csv('../zoo/class.csv')

# 1. Total unique animal species
# Since 'Animal_List' already contains lists, we can directly explode the lists and count unique species
df['Animal_List'] = df['Animal_Names'].apply(lambda x: x.split(', '))
unique_animals = set([animal for sublist in df['Animal_List'] for animal in sublist])
print(f"Total unique animal species: {len(unique_animals)}")

# Visualization for unique animal species
plt.figure(figsize=(10, 6))
df['Animal_List'].explode().value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Most Common Animal Species')
plt.ylabel('Frequency')
plt.show()

# 2. Average number of species per class
# Calculate the mean number of species across all classes
average_species = df['Number_Of_Animal_Species_In_Class'].mean()
print(f"Average number of species per class: {average_species}")

# Visualization for average number of species per class
plt.figure(figsize=(10, 6))
df['Number_Of_Animal_Species_In_Class'].plot(kind='hist', bins=10, color='skyblue')
plt.title('Distribution of Number of Species per Class')
plt.xlabel('Number of Species')
plt.ylabel('Frequency')
plt.show()

# 3. Class with the most animal species
# Find the class with the maximum number of species
max_species_class = df.loc[df['Number_Of_Animal_Species_In_Class'].idxmax()]
print(f"Class with most animal species: {max_species_class['Class_Type']} with {max_species_class['Number_Of_Animal_Species_In_Class']} species")

# Visualization for class with the most species
plt.figure(figsize=(10, 6))
df.plot(kind='bar', x='Class_Type', y='Number_Of_Animal_Species_In_Class', color='green')
plt.title('Number of Animal Species per Class')
plt.xlabel('Class')
plt.ylabel('Number of Species')
plt.show()

# 4. Class with the fewest animal species
# Find the class with the fewest number of species
min_species_class = df.loc[df['Number_Of_Animal_Species_In_Class'].idxmin()]
print(f"Class with fewest animal species: {min_species_class['Class_Type']} with {min_species_class['Number_Of_Animal_Species_In_Class']} species")

# Visualization for class with the fewest species
plt.figure(figsize=(10, 6))
df.plot(kind='bar', x='Class_Type', y='Number_Of_Animal_Species_In_Class', color='red')
plt.title('Number of Animal Species per Class')
plt.xlabel('Class')
plt.ylabel('Number of Species')
plt.show()

# 5. Total number of mammal species
# Find the number of mammal species
mammals = df[df['Class_Type'] == 'Mammal']
mammal_species = mammals['Number_Of_Animal_Species_In_Class'].sum()
print(f"Total number of mammal species: {mammal_species}")

# Visualization for total number of mammal species
plt.figure(figsize=(10, 6))
mammals.plot(kind='bar', x='Class_Type', y='Number_Of_Animal_Species_In_Class', color='blue')
plt.title('Mammal Species Count')
plt.xlabel('Class')
plt.ylabel('Number of Species')
plt.show()
```
